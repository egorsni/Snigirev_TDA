```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.linear = nn.Linear(9, 9, bias=False) # bias это отступ AX + b, b -- bias
        self.W = nn.Linear(45, 50, bias=False)
        
    def forward(self, x):
#         D = torch.Tensor()
#         for direction_data in x:
#             H = self.linear(torch.Tensor(direction_data).T).T
#             H /= H.shape[1]
#             dn = torch.sum(H, dim = 1)
# #             print((self.linear(torch.Tensor(direction_data).T).T).shape, H.shape[1])
# #             D.append(dn)
#             D = torch.cat((D, dn),dim=0)
        x = x.reshape((9,-1))
        D = self.linear(torch.Tensor(x).T).T
        D = torch.reshape(D,(-1,9))
#         print(D)
        D = D.T
#         return torch.softmax(self.W((torch.reshape(torch.softmax(D @ torch.transpose(D, 0, 1), -1) @ D, (-1,1))).T).T, 0)
        probs = self.W((torch.reshape(torch.softmax(D @ torch.transpose(D, 0, 1), -1) @ D, (-1,1))).T).T.reshape((5,10))
        return probs
```

```python
model = NeuralNetwork()
print(model)
```

```python
learning_rate = 1e-3
batch_size = 5
epochs = 5

loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
from torch.utils.data import TensorDataset, DataLoader
tensor_x = torch.Tensor(X_train) # transform to torch tensor
tensor_y = torch.Tensor(y_train)
print(tensor_x.shape)

my_dataset = TensorDataset(tensor_x,tensor_y) # create your datset
train_dataloader = DataLoader(my_dataset,batch_size=batch_size, shuffle=True) # create your dataloader

tensor_x = torch.Tensor(X_test) # transform to torch tensor
tensor_y = torch.Tensor(y_test)

my_dataset = TensorDataset(tensor_x,tensor_y) # create your datset
test_dataloader = DataLoader(my_dataset,batch_size=batch_size, shuffle=True) # create your dataloader
```

```python
def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    for batch, (X, y) in enumerate(dataloader):
        # Compute prediction and loss
        pred = model(X)
#         print(pred.shape, y.shape)
#         print(pred, y)
        loss = loss_fn(pred, y.type(torch.LongTensor))

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), batch * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")


def test_loop(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0

    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y.type(torch.LongTensor)).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()

    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

epochs = 5
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loop(train_dataloader, model, loss_fn, optimizer)
    test_loop(test_dataloader, model, loss_fn)
print("Done!")
```
